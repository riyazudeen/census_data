```python
import pandas as pd

df = pd.read_excel('./census_2011.xlsx',header=0)
df.head()
```

```python
df = df.rename(columns = { 
                            'State name' : 'State/UT','District name':'District',
                            'Male_Literate':'Literate_Male','Female_Literate':'Literate_Female',
                            'Rural_Households': 'Households_Rural','Urban_ Households' : 'Households_Urban',
                            'Age_Group_0_29': 'Young_and_Adult','Age_Group_30_49': 'Middle_Aged',
                            'Age_Group_50': 'Senior_Citizen', 'Age not stated': 'Age_Not_Stated'
                        })

df.head()
```

```python
import re

df['State/UT'] =  df['State/UT'].str.title()

def isContainAnd(state):
    if state != "Andhra Pradesh" :
        if "And" in state:
            partten = r"\bAnd\b"
            new_value = re.sub(partten, "and", state)
            return  new_value
        else:
            return state
    else:
        return state

df['State/UT'] = df['State/UT'].apply(isContainAnd)

df.head()
```

```python
import docx

def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return fullText


telanganaDistric = getText(r'D:\capstone1\Telangana.docx')

print(telanganaDistric)
```

```python
def setTelangana():
    for  i in telanganaDistric :
        df.loc[ df['District'] == i, 'State/UT'] ='Telangana'

setTelangana()

ladakhDistric= [
    "Leh(Ladakh)",
    "Kargil"
]

def setLadakh():
    for  i in ladakhDistric :
        df.loc[ df['District'] == i, 'State/UT'] ='Ladakh'


setLadakh()

df[df['State/UT'].str.contains('Ladakh')]
```

```python

df['Population'].fillna(value=df['Male']+df['Female'],inplace=True)
df['Female'].fillna(value=df['Population'] - df['Male'], inplace=True)
df['Male'].fillna(value=df['Population'] - df['Female'], inplace=True)
df['Literate'].fillna(value=df['Literate_Male']+df['Literate_Female'],inplace=True)
df['Literate_Male'].fillna(value=df['Literate'] - df['Literate_Female'], inplace=True)
df['Literate_Female'].fillna(value=df['Literate'] - df['Literate_Male'], inplace=True)
df['ST'].fillna(value=df['Male_ST']+df['Female_ST'],inplace=True)
df['Male_ST'].fillna(value=df['ST'] -df['Female_ST'], inplace=True)
df['Female_ST'].fillna(value=df['ST'] -df['Male_ST'], inplace=True)
df['Workers'].fillna(value=df['Male_Workers']+df['Female_Workers'],inplace=True)
df['Male_Workers'].fillna(value=df['Workers'] -df['Female_Workers'], inplace=True)
df['Female_Workers'].fillna(value=df['Workers'] -df['Male_Workers'], inplace=True)
```

```python
def creatdp(df):
    mycursor.execute('CREATE DATABASE IF NOT EXISTS census_db')

    listOfColumns = list()

    for i in df.columns:
        if i == 'District code':
            listOfColumns.append('District_Code INT PRIMARY KEY')
        elif i == "State/UT":
            listOfColumns.append('State_UT VARCHAR(50)')
        elif i == "District":
            listOfColumns.append('District VARCHAR(50)')
        elif i == "Households_with_TV_Computer_Laptop_Telephone_mobile_phone_and_Scooter_Car":
            listOfColumns.append("Households_TV_Computer_Laptop_Telephone_mobile_phone_Scooter_Car INT")
        elif i == "Type_of_latrine_facility_Night_soil_disposed_into_open_drain_Households":
            listOfColumns.append("Type_of_latrine_facility_Night_soil_disposed_into_open_drain INT")
        elif i == "Type_of_latrine_facility_Flush_pour_flush_latrine_connected_to_other_system_Households":
            listOfColumns.append("Type_of_latrine_Flush_pour_connected_to_other_system_Households INT")
        elif i == "Not_having_latrine_facility_within_the_premises_Alternative_source_Open_Households":
            listOfColumns.append("Not_having_latrine_within_premises_Other_source_Open_Households INT")
        elif i == "Main_source_of_drinking_water_Handpump_Tubewell_Borewell_Households":
            listOfColumns.append("Source_of_drinking_water_Handpump_Tubewell_Borewell_Households INT")
        elif i == "Main_source_of_drinking_water_Other_sources_Spring_River_Canal_Tank_Pond_Lake_Other_sources__Households":
            listOfColumns.append("Drinking_water_Spring_River_Canal_Tank_Pond_Lake_Other_Household INT")
        else:
            listOfColumns.append(f'{i} INT')
    
    mycursor.execute(f'CREATE TABLE census_db.census_2011 ({", ".join(listOfColumns)})')
    mydb.commit()

creatdp(df)
```

```python
from bson import ObjectId  
myCollection = client['test']
col = myCollection['census']
#mycursor.execute("USE test")

# def create_database(mysql_cursor, database_name):
#     try:
#         mysql_cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database_name}")
#         print(f"Database '{database_name}' created (if it didn't exist).")
#     except mysql.connector.Error as err:
#         print(f"Error creating database: {err}")


def prepare_data_for_mysql(mongo_data):
    mysql_data = {}
    for key, value in mongo_data.items():
        if key == "_id":  # Handle MongoDB's _id field (optional)
            continue
        # Convert data types (e.g., BSON ObjectId to string)
        if isinstance(value, ObjectId):
            value = str(value)
        mysql_data[key] = value  # No manual escaping needed with prepared statements
    return mysql_data

def insert_data_into_mysql(mysql_cursor, table_name, data):
    columns = ", ".join(data.keys())
    placeholders = ", ".join(["%s"] * len(data))
    query = f"INSERT INTO census_db.{table_name} ({columns}) VALUES ({placeholders})"
    try:
        mysql_cursor.execute(query, tuple(data.values()))
        mydb.commit()
        print(f"Successfully inserted data into '{table_name}'.")
    except mysql.connector.Error as err:
        print(f"Error inserting data: {err}")
        mydb.rollback()
        
# create_database(mycursor, "census_db") 
       
for mongo_document in col.find():    
    mysql_data = prepare_data_for_mysql(mongo_document)
    insert_data_into_mysql(mycursor, "census_2011", mysql_data)

#client.close()
mycursor.close()
mydb.close()

```
